```python
import pandas as pd
from sklearn.datasets import make_blobs

data = make_blobs(n_samples=1000, centers=3, cluster_std=1.5, center_box=(-6.0, 6.0), random_state=62)[0]
df = pd.DataFrame(data, columns=['x1', 'x2'])
df.head()
```

```python
import plotly.express as px

px.scatter(df, 'x1', 'x2', width=950, height=500, title='Dane')
```

```python
from sklearn.cluster import KMeans
km_model = KMeans(n_clusters=1)
```

```python
km_model.fit(data)
```

```python
y_kmeans = km_model.predict(data)
```

```python
df['cluster']=y_kmeans
df.head()
```

```python
import plotly.graph_objects as go

centroid_1, = km_model.cluster_centers_

fig = px.scatter(df, 'x1', 'x2', 'cluster', width=950, height=500, title='Algorytm K-średnich - 1 klastry')
fig.add_trace(go.Scatter(x=[centroid_1[0]], y=[centroid_1[1]], name='centroid 1', mode='markers', marker_line_width=2))
fig
```

```python
km_model = KMeans(n_clusters=2)
km_model.fit(data)
y_kmeans = km_model.predict(data)
df['cluster']=y_kmeans
```

```python
import plotly.graph_objects as go

centroid_1, centroid_2 = km_model.cluster_centers_

fig = px.scatter(df, 'x1', 'x2', 'cluster', width=950, height=500, title='Algorytm K-średnich - 2 klastry')
fig.add_trace(go.Scatter(x=[centroid_1[0]], y=[centroid_1[1]], name='centroid 1', mode='markers', marker_line_width=2))
fig.add_trace(go.Scatter(x=[centroid_2[0]], y=[centroid_2[1]], name='centroid 2', mode='markers', marker_line_width=2))
fig
```

```python
km_model = KMeans(n_clusters=3)
km_model.fit(data)
y_kmeans = km_model.predict(data)
df['cluster']=y_kmeans
```
